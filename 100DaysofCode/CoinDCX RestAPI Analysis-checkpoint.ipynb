```python
import pandas as pd
import requests

# Fetch JSON data from the URL
url = "https://api.coindcx.com/exchange/v1/markets_details"
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    data = response.json()  # Convert the response to JSON format

    # Convert JSON to DataFrame
    df = pd.DataFrame(data)

    # Display DataFrame
    print(df)
else:
    print("Failed to fetch data from the URL")
```

```python
df.head(5)
```

```python
df2 = df[['coindcx_name','target_currency_short_name','max_quantity','min_quantity']]
df2
```

```python
filtered_data = df2[df2['max_quantity']<1000.0]
filtered_data
```

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by 'date' for proper visualization
filtered_data = filtered_data.sort_values('max_quantity')

plt.figure(figsize=(10,6))

plt.plot(filtered_data['max_quantity'],filtered_data['min_quantity'],marker='o',linestyle='-')
plt.xlabel('Max_Quantity of Cryptocurrencies')
plt.ylabel('Min_Quantity of Cryptocurrencies')
plt.title('Variation between Max and Min quantities of Cryptocurrencies')


# Display the plot
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```
